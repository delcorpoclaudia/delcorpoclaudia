```python
from sklearn.cluster import KMeans
import sklearn.datasets
import matplotlib.pyplot as plt
import numpy as np

# Random dataset
data = sklearn.datasets.make_blobs(n_samples=100, n_features=2)[0]


# Initialize KMeans with 3 clusters
kmeans = KMeans(n_clusters=3)

# Fit the model
kmeans.fit(data)

# Predict the clusters
labels = kmeans.predict(data)

# Visualize the clusters
plt.scatter(data[:,0], data[:,1], c=labels, cmap='rainbow')
plt.title('K-Means Clustering')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
data = sklearn.datasets.make_blobs(n_samples=100, n_features=2, centers=5)[0]

# Initialize DBSCAN
dbscan = DBSCAN(eps=1.0, min_samples=5)

# Fit the model
labels = dbscan.fit_predict(data)
print(labels)

# Visualize the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='rainbow')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 400)
y = sigmoid(x)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$f(x) = \frac{1}{1 + e^{-x}}$', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of the sigmoid function
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Generate x values
x = np.linspace(-10, 10, 400)
y = sigmoid(x)
y_prime = sigmoid_derivative(x)

# Plot both sigmoid and its derivative
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'Sigmoid: $f(x) = \frac{1}{1 + e^{-x}}$', color='blue')
plt.plot(x, y_prime, label=r"Derivative: $f'(x) = f(x)(1 - f(x))$", color='red', linestyle='--')
plt.title('Sigmoid Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the ReLU function
def relu(x):
    return np.maximum(0, x)

# Define the derivative of ReLU
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Generate x values
x = np.linspace(-10, 10, 400)
y = relu(x)
y_prime = relu_derivative(x)

# Plot both ReLU and its derivative
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='ReLU: $f(x) = \max(0, x)$', color='green')
plt.plot(x, y_prime, label="Derivative: $f'(x)$", color='orange', linestyle='--')
plt.title('ReLU Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```
